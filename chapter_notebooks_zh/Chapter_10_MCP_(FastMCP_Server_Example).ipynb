```
# fastmcp_server.py
# 该脚本演示了如何使用 FastMCP 创建简单的 MCP 服务器。
# 它公开了一个生成问候语的工具。

# 要运行此服务器：
# 1. 确保已安装 FastMCP： pip install fastmcp
# 2.将此代码保存为fastmcp_server.py
# 3. 从终端运行：python fastmcp_server.py

from fastmcp import FastMCP, tool
import asyncio # FastMCP 异步功能所需

# 定义一个简单的工具函数。
# `@tool()` 装饰器将此 Python 函数注册为 MCP 工具。
# 文档字符串成为 LLM 工具的描述。
@tool()
def greet(name: str) -> str:
    """
    Generates a personalized greeting.

    Args:
        name: The name of the person to greet.

    Returns:
        A greeting string.
    """
    return f"Hello, {name}! Nice to meet you."

# 初始化 FastMCP 服务器。
# 默认情况下，FastMCP 在 http://localhost:8000 上运行
# 并自动发现用@tool()修饰的函数。
mcp_server = FastMCP()

# 要运行服务器，通常使用“mcp_server.run()”或“mcp_server.run_async()”。
# 对于一个简单的脚本，“run()”就足够了。
if __name__ == "__main__":
    print("Starting FastMCP server...")
    print("This server exposes a 'greet' tool.")
    print("Access the tool schema at http://localhost:8000/tools.json")
    print("Press Ctrl+C to stop the server.")
    # FastMCP 的 run() 方法阻塞并启动服务器。
    # 它在内部处理异步事件循环。
    mcp_server.run()
```
